```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LinearRegression, LogisticRegression
# Generate data 
x, y = make_classification(n_samples=100, n_features=1, n_informative=1, n_redundant=0, 
                           n_classes=2, n_clusters_per_class=1, class_sep=0.25, random_state=0)
model = LinearRegression()
model.fit(x,y)
inputs = [[x/100] for x in range(-100,100)]
predictions = model.predict(inputs)
modelLR = LogisticRegression()
modelLR.fit(x,y)
predictionsLR = modelLR.predict(inputs)
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(x, y, s=50, alpha=0.5)
plt.xlabel('feature')
plt.ylabel('target')
plt.plot(inputs,predictions)
plt.title('line ear reeee gressshun')
plt.plot(inputs,predictionsLR)
temp = modelLR.predict_proba(inputs)[:,1]
```
