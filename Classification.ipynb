```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LinearRegression, LogisticRegression
#Generate rando date-o
x,y = make_classification(n_samples=100, n_features=1, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, class_sep=0.25, random_state=0)
```

```python
linearmodel = LinearRegression()
linearmodel.fit(x,y)
inputs = [[x/100] for x in range(-100,100)]
predictions = linearmodel.predict(inputs)
```

```python
logisticmodel = LogisticRegression()
logisticmodel.fit(x,y)
logpredicitons = logisticmodel.predict(inputs)
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(x,y, s=50, alpha=0.5)
plt.xlabel('feature')
plt.ylabel('target')

#plt.plot(inputs, predictions);
plt.plot(inputs,logpredicitons);
```
